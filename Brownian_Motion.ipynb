#####Question 1: Simulate Standard Brownian Motion

```
from math import sqrt
from scipy.stats import norm
import numpy as np


def brownian_motion(x_init, n, dt, delta, out=None):
    x_init = np.asarray(x_init)

    # For each element of x_init, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x_init.shape + (n,), scale=delta*sqrt(dt))

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)

    # Add the initial condition.
    out += np.expand_dims(x_init, axis=-1)

    return out
```

# Wiener Process in 1D

```
import numpy
from numpy import random
from pylab import plot, show, grid, xlabel, ylabel

# The Wiener process parameter.
delta = 2
# Total time.
T = 15.0
# Number of steps.
N = 800
# Time step size
dt = T/N
# Number of realizations to generate.
m = 1
# Create an empty array to store the realizations.
x = numpy.empty((m,N+1))
# Initial values of x.
x[:, 0] = random.randint(100)

brownian_motion(x[:,0], N, dt, delta, out=x[:,1:])

t = numpy.linspace(0.0, N*dt, N+1)
for k in range(m):
    plot(t, x[k])
xlabel('t', fontsize=16)
ylabel('x', fontsize=16)
grid(True)
show()
```
